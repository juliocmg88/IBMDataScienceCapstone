```python
# TASK 3: Calculate the distances between a launch site to its proximities

from geopy.distance import geodesic
proximities = {
    'Miami Airport': (25.7959, -80.2870),
    'Los Angeles Airport': (33.9416, -118.4085),
    'Houston Downtown': (29.7604, -95.3698),
    'Pacific Ocean Coastline': (28.396837, -80.605659),  # Close to Cape Canaveral
}

# Get distinct launch sites and their coordinates
launch_sites = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates()

# Loop through launch sites and calculate distances
for _, site in launch_sites.iterrows():
    site_name = site['Launch Site']
    site_coord = (site['Lat'], site['Long'])
    print(f"\nLaunch Site: {site_name}")
    
    for place, coord in proximities.items():
        dist = geodesic(site_coord, coord).km
        print(f"  → Distance to {place}: {dist:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

# Función para calcular distancia geodésica
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radio de la tierra en km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Coordenadas del sitio de lanzamiento
launch_site = [28.562302, -80.577356]

# Coordenadas de puntos cercanos
closest_city = [28.3852, -80.6081]
closest_highway = [28.5721, -80.6480]
closest_railway = [28.5234, -80.6725]

# Crear el mapa
site_map = folium.Map(location=launch_site, zoom_start=11)

# Agregar marcador del sitio de lanzamiento
folium.Marker(launch_site, popup='Launch Site', icon=folium.Icon(color='blue')).add_to(site_map)

# Lista de puntos y etiquetas
proximities = {
    'Closest City': closest_city,
    'Closest Highway': closest_highway,
    'Closest Railway': closest_railway
}

# Agregar marcadores, líneas y etiquetas de distancia
for label, coord in proximities.items():
    # Marcar el punto
    folium.Marker(coord, popup=label, icon=folium.Icon(color='green')).add_to(site_map)

    # Dibujar línea
    folium.PolyLine(locations=[launch_site, coord], color='gray', weight=2).add_to(site_map)

    # Calcular distancia
    dist_km = calculate_distance(launch_site[0], launch_site[1], coord[0], coord[1])

    # Añadir marcador con texto de distancia en el punto medio
    mid_lat = (launch_site[0] + coord[0]) / 2
    mid_lon = (launch_site[1] + coord[1]) / 2
    folium.Marker(
        [mid_lat, mid_lon],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color: #d35400;"><b>{label}: {dist_km:.2f} km</b></div>',
        )
    ).add_to(site_map)

# Mostrar el mapa
site_map
```
